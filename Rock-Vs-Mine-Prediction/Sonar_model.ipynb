```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Processing

```python
sonar_data = pd.read_csv('/content/Sonar_data.csv', header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data[60].value_counts()

# here we got almost similar numbers for both the categories so we don't have to stratify the data compulsarly
# M    111
# R     97
# Name: 60, dtype: int64
```

# Separating label and features

```python
X = sonar_data.drop(60, axis=1)
# (shape= 208, 60)
# (if we didn't specified axis above, shape would have (207, 61) by dropping 60th row)
X
```

```python
y = sonar_data[60]
y.head()
# (m is mine and r is rock)
```

```python
# a simple explanation of how groupby works:

# Splitting: The DataFrame is split into groups based on one or more criteria.
# Applying: A function is applied to each group independently.
# Combining: The results of the function applications are combined back into a new DataFrame.

# data = {'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
#         'Value': [10, 20, 30, 40, 50, 60]}

# df = pd.DataFrame(data)

# # Grouping by 'Category'
# grouped = df.groupby('Category')

# # Calculating the mean for each group
# mean_values = grouped.mean()

# print(mean_values)
```

# Splitting training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
len(x_test)
```

```python
len(x_train)
```

# Model training

```python
model = LogisticRegression()
```

Traing our loogistic regression model

```python
model.fit(x_train, y_train)

# The .fit() method takes the feature matrix X_train and target variable y_train as arguments and fits
# the logistic regression model to the training data. During this process, the model learns the coefficients
# and intercept that define the decision boundary.

# After the model is trained using .fit(), we can use the trained model to make predictions on new data using
# the .predict() method.
```

# Model Evaluation

```python
# let's see the accuracy on training data first
x_train_prediction = model.predict(x_train)
```

```python
x_train_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on training data :', x_train_accuracy)
# Accuracy on training data : 0.8373493975903614
```

```python
# Accuracy on test data
x_test_prediction = model.predict(x_test)
x_test_accuracy = accuracy_score(x_test_prediction, y_test)
print('Accuracy on testing data :', x_test_accuracy)
# Accuracy on testing data : 0.8571428571428571
# Don't know how i'm getting better prediction in my test data than train data 🤷‍♀️
```
