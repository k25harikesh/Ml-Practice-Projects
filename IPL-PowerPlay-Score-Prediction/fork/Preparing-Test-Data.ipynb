```python
import pandas as pd
```

### Making a dataframe from given feature and labels

```python
dfs = []

for i in range(12, 71):
# 46 test data was missing)
    if i == 46:
        continue
    
# reading labels and features
    labels_file = f'../Data/FilesUsed/test_file_labels_matchid_{i}.csv'    
    features_file = f'../Data/FilesUsed/test_file_matchid_{i}.csv'
    
    labels = pd.read_csv(labels_file)
    features = pd.read_csv(features_file)
    
# features.iloc[:, 0]  get all the rows of 1st column

# we removed the first column of features as it was worthless after seeing all the datas
    features =features.iloc[:, 1:]

# [:, 1:] is using slicing to select all rows (:) and all columns starting from the second column (1:) onward.

# (features = features.iloc[1:])
#  So, this line of code is essentially removing the first row from the DataFrame features and assigning the result back to the variable features.
# The resulting DataFrame will start from the second row and include all subsequent rows.

# Add 'total_runs' column from labels
    features['total_runs'] = labels['actual_runs']
    
    # Insert 'match_id' column at the beginning
    features.insert(loc=0, column='match_id', value=i)

    dfs.append(features)
#   dfs is a list of multiple datframes but not a datafrme so make a combined dataframe
```

```python
dfs
```

```python
combined_df = pd.concat(dfs, axis=0, ignore_index=True)
# concantinating multiple dataframes which were packed in list along the rows(vertical stacking as axis=0) with fresh index
```
