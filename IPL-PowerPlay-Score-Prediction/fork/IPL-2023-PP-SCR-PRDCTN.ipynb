## Data Cleaning (Removing unwanted columns)

```python
# Renaming columns for ball_by_ball DataFrame
ball_by_ball = ball_by_ball_orig.rename(columns={
    'ID': 'match_id',
    'ballnumber': 'ball_number',
    'non-striker': 'non_striker',
    'BattingTeam': 'batting_team',
})

# selecting columns for ball_by_ball DataFrame
ball_by_ball = ball_by_ball.loc[:, [
    'match_id',
    'innings',
    'batting_team',
    'overs',
    'ball_number',
    'batter',
    'bowler',
    'total_run',
]]
# The loc method is used to select specific columns in the DataFrame using the labels.
# final DataFrame ball_by_ball is a subset of the original DataFrame ball_by_ball_orig,
# containing only the specified columns with the renamed column names.

# Renaming and selecting columns for matches_result DataFrame
matches_result = matches_result_orig.rename(columns={
    'ID': 'match_id',
    'Team1': 'team_1',
    'Team2': 'team_2',
    'Venue': 'venue',
}).loc[:, [
    'match_id',
    'team_1',
    'team_2',
    'venue',
]]
```

## Applying team and venue mapping

```python
matches_result.venue = matches_result['venue'] =\
matches_result.venue.map(venue_mapping_normal).fillna('Other_ven')

# matches_result.venue.map(venue_mapping_normal): This maps values in the 'venue' column 
# using the venue_mapping_normal function or dictionary.
# It replaces each value in the 'venue' column with its corresponding mapped value.
```

```python
matches_result.team_1 = matches_result.team_1.map(team_mapping).fillna('Other')
matches_result.team_2 = matches_result.team_2.map(team_mapping).fillna('Other')
```

```python
ball_by_ball.batting_team = ball_by_ball.batting_team.map(team_mapping).fillna('Other')
```
